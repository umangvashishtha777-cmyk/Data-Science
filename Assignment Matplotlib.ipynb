```python
import matplotlib.pyplot as plt
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
kohli = [995, 1381, 1026, 1268, 1054, 1380, 2595, 2818, 2735, 2455]
rohit = [606, 928, 777, 837, 864, 815, 1240, 1293, 1030, 1490]
sehwag = [785, 505, 503, 215, 112, 0, 0, 0, 0, 0]

plt.plot(years,kohli,label='kohli')
plt.plot(years,rohit,label='rohit')
plt.plot(years,sehwag,label='sehwag')

plt.title("Run's Comparison")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli,color='yellow',label='kohli')
plt.plot(years,rohit,color='green',label='rohit')
plt.plot(years,sehwag,color='red',label='sehwag')

plt.title("Run's Comparison")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli,color='yellow',linestyle='--',label='kohli')
plt.plot(years,rohit,color='green',linestyle=':',label='rohit')
plt.plot(years,sehwag,color='red',linestyle='-.',label='sehwag')

plt.title("Run's Comparison")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(years,kohli,color='yellow',linestyle='--',label='kohli')
plt.plot(years,rohit,color='green',linestyle=':',label='rohit')
plt.plot(years,sehwag,color='red',linestyle='-.',label='sehwag')

plt.title("Run's Comparison")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.show()
```
