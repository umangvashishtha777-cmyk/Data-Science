```python
import numpy as np
import time
```

```python


start = time.time()
arr1 = np.array([1, 2, 3, 4, 5])
list_ans = arr1 ** 2
end = time.time()

rest = end - start
print("Squared Array:", list_ans)
print("Time taken:", rest)
```
